```python
import pandas as pd
from padelpy import padeldescriptor, from_smiles
from rdkit import Chem
from os import remove
```

### 2. Cardiotoxicity Pubchem fingerprints

```python
fingerprint_output_file = 'data/raw/cardio_pubchem.csv'
fingerprint_descriptortypes = 'PubchemFingerprinter.xml'
mol_dir='data/raw/hERG_all_human_data_filtered_11504cmds.sdf'

padeldescriptor(mol_dir=mol_dir, 
                d_file=fingerprint_output_file, 
                descriptortypes= fingerprint_descriptortypes,
                detectaromaticity=True,
                standardizenitro=True,
                standardizetautomers=True,
                threads=2,
                removesalt=True,
                fingerprints=True)
```

```python
fingerprints = pd.read_csv(fingerprint_output_file)
fingerprints
```
